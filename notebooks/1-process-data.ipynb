```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/PubChem_compound_text_propane.csv")
print(len(df))
df.head()
```

```python
df.columns
```

# Feature Engineering

## Drop Missing IUPAC

```python
df.dropna(subset='iupacname', inplace=True)
```

## Remove Multiple compound

```python
df_select = df[~df['iupacname'].str.contains(';')]

df_select
```

```python
df_select[['cid', 'iupacname']]
```

## Remove Duplicate IUPAC

```python
df_select.drop_duplicates(subset='iupacname', inplace=True)
```

```python
df_select[['complexity', 'cid', 'iupacname']]
```

## Remove Isotope C

```python
df_output = df_select[~df_select['iupacname'].str.contains("C\)")].reset_index(drop=True)
```

```python
df_output[['cid', 'complexity', 'iupacname']]
```

# Export Transformed

```python
df_output[['cid', 'iupacname']].to_csv("../data/interim/propane_iupac.csv", index=False)
```
